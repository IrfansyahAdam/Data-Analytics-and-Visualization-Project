```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['TGLMASUK']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'TGLMASUK'}, axis=1)
              .sort_values('TGLMASUK', ascending=True))
  xs = counted['TGLMASUK']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(15, 5.2), layout='constrained')
df_sorted = df_sheet1.sort_values('TGLMASUK', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('NAMAJALUR')):
  _plot_series(series, series_name, i)
  fig.legend(title='Nama Jalur', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Tahun Masuk')
_ = plt.ylabel('count()')
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# File path of the excel file in your Google Drive
file_path = '/content/drive/My Drive/Dataset-Mhs-Pendfisika.xlsx'

# Read the excel file using Pandas
df_sheet1 = pd.read_excel(file_path, sheet_name='Data-mahasiswa')
df_sheet2 = pd.read_excel(file_path, sheet_name='Data-akademis')

# Grouping the data by 'TGLMASUK' and 'TGLLULUS"
perbandingan_tanggal = df_sheet1.groupby('TGLMASUK')['TGLLULUS'].value_counts()

# Creating a graphic chart
plt.figure(figsize=(20, 8))
perbandingan_tanggal.plot(kind='line', marker='o', color='red')
plt.xlabel('Periode Kuliah')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Grafik perbandingan Periode kuliah dengan jumlah mahasiswa')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['ANGKATAN']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'ANGKATAN'}, axis=1)
              .sort_values('ANGKATAN', ascending=True))
  xs = counted['ANGKATAN']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df_sheet2.sort_values('ANGKATAN', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('ANGKATAN')
_ = plt.ylabel('count()')
```
