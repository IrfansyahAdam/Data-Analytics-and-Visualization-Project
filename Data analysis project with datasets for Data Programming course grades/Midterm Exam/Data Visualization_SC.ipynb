```python
print ('-'*70)
print ('\tTM UTS  \nNama    \t: Muhammad Irfansyah Adam \nNRM     \t: 1306622055 \nKelas   \t: FISIKA A')
print ('-'*70)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Contoh data
data = pd.read_excel("/content/drive/MyDrive/TM UTS/survey.xlsx")
# Pastikan kolom 'Age' dan 'Streaming_service' ada dalam DataFrame 'data'

# Menghitung rata-rata umur untuk setiap Streaming Service
average_age_by_service = data.groupby('Streaming_service')['Age'].mean().reset_index()

# Membuat bar plot
plt.figure(figsize=(12, 6))  # Mengatur ukuran grafik
ax = sns.barplot(data=average_age_by_service, x='Age', y='Streaming_service', palette='pastel')

plt.xlabel('Rata-rata Umur')
plt.ylabel('Streaming Service')
plt.title('Rata-rata Umur pengguna Streaming Service')

# Menambahkan label data pada setiap bar
for p in ax.patches:
    width = p.get_width()
    plt.text(width, p.get_y() + p.get_height() / 2, f'{width:.2f}', ha="left")

plt.show()

```
