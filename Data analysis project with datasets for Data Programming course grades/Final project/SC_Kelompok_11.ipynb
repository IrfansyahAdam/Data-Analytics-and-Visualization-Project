```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# File path of the CSV file in your Google Drive
file_path = '/content/drive/My Drive/Datasettaman.csv'

# Read the CSV file using Pandas
data = pd.read_csv(file_path)

# Count the frequency of each unique value in the 'kotamadya' column
kotamadya_frequency = data['kotamadya'].value_counts()

# Creating a bar chart for frequency of 'kotamadya' values
plt.figure(figsize=(10, 6))
kotamadya_frequency.plot(kind='bar', color='skyblue')
plt.xlabel('Kotamadya')
plt.ylabel('Jumlah variasi taman')
plt.title('Jumlah variasi taman dalam Kotamadya')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# File path of the CSV file in your Google Drive
file_path = '/content/drive/My Drive/Datasettaman.csv'

# Read the CSV file using Pandas
data = pd.read_csv(file_path)

# Grouping the data by 'kotamadya' and calculating the sum of 'luas' for each 'kotamadya'
luas_sum_by_location = data.groupby('kotamadya')['luas'].sum()

# Creating a line chart
plt.figure(figsize=(10, 6))
luas_sum_by_location.plot(kind='line', marker='o', color='red')
plt.xlabel('Kotamadya')
plt.ylabel('Jumlah luas dalam meter persegi')
plt.title('Grafik luas taman dalam Kotamadya')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# File path of the CSV file in your Google Drive
file_path = '/content/drive/My Drive/Datasettaman.csv'

# Read the CSV file using Pandas
data = pd.read_csv(file_path)

# Count the frequency of each unique value in the 'Kotamadya' column
kotamadya_frequency = data['kotamadya'].value_counts()

# Creating a pie chart for frequency of 'Kotamadya' values
plt.figure(figsize=(8, 8))
plt.pie(kotamadya_frequency, labels=kotamadya_frequency.index, autopct='%1.1f%%', startangle=140)
plt.title('Jumlah taman yang ada berdasarkan kotamadya')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# File path of the CSV file in your Google Drive
file_path = '/content/drive/My Drive/Datasettaman.csv'

# Read the CSV file using Pandas
data = pd.read_csv(file_path)

# Calculate the sum of 'luas' for each unique 'kotamadya'
luas_sum_by_kotamadya = data.groupby('kotamadya')['luas'].sum()

# Creating a pie chart for the sum of 'luas' by 'kotamadya'
plt.figure(figsize=(8, 8))
plt.pie(luas_sum_by_kotamadya, labels=luas_sum_by_kotamadya.index, autopct='%1.1f%%', startangle=140)
plt.title('Jumlah luas total taman dalam kotamadya')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```
